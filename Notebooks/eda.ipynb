```python
vehicles_data['model_year'] = vehicles_data.groupby('model')['model_year'].transform(lambda x: x.fillna(x.median()))
vehicles_data['cylinders'] = vehicles_data.groupby('model')['cylinders'].transform(lambda x: x.fillna(x.median()))
vehicles_data['odometer'] = vehicles_data.groupby(['model_year', 'model'])['odometer'].transform(lambda x: x.fillna(x.median()))

# Here I have removed an NaN's with there median value for the each column, model_year, cylinders and odometers (however odometer Im having trouble with it doesnt seem to be removing the NaN's)
vehicles_data.info()
```

```python
def remove_outliers(vehilces_data, column):
    Q1 = vehicles_data[column].quantile(0.25)
    Q3 = vehicles_data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return vehicles_data[(vehicles_data[column] >= lower_bound) & (vehicles_data[column] <= upper_bound)]

vehicles_data = remove_outliers(vehicles_data, 'model_year')
vehicles_data = remove_outliers(vehicles_data, 'price')

# Here I am removing outliers from both the model_year column as well as the price column and we are doing this due to the great difference in the median and mean we saw above. 
# The function uses a method involving quartiles and the Interquartile range (IQR) to identify outliers. 
    # Q1 is the value below which 25% of the data falls.
    # Q2 is the value below which 75% of the data falls. i.e finding the cars that are cheaper than 75% of the cars, leaving the top 25% as the ost expensive
    # IQR is the difference between Q3 and Q1. It's a measure of where the "middle half" of my data is. Representing the range of prices for the middle 50% of the cars, 
        # ignoring the cheapest 25% and the most expenseive 25%
        
    # Multiplying the IQR by 1.5 gives me a way to define what "too unusual" means in terms of distance from the middle 50%. The number 1.5 is arbitray but widely accepted for this purpose. 
        # It strikes a blance that captures the most "normal" data while excluding extreme outliers 
    # The return line essentially is telling Python to give me back only the cars that fall within the defined 'normal range'. 
        # It takes all the cars that have a price or model_year between the lower and upper boundaries I calculated. Cars outside these bounaries are considered outliers and therefor are not returned 
        
```

```python
pivot_df = vehicle_type_counts.pivot(index='manufacturer', columns='type', values= 'Number of Vehicles').fillna(0)
# Here we are using are previous vehicle counts data which found the number of vehicles that each manufactuer has in each vehicle type for this dealership
pivot_df
```

```python
melted_df = pivot_df.reset_index().melt(id_vars='manufacturer', var_name='type', value_name='Number of Vehicles')

  # Created the bar graph with Plotly Express
fig = px.bar(melted_df,
             x='manufacturer',
             y='Number of Vehicles',
             color='type',  # Differentiates vehicle types with color
             title='Distribution of Vehicle Types by Manufacturer')



# Rotated x-axis labels for better readability
fig.update_xaxes(tickangle=45)


fig.show()
```

By pivoting the data each manufacturer is an index, vehicle types are columns, and the values are the counts of vehicles for each type. Missing values are filled with 0 indicating that the manufacturer does not have that vehicle type listed. 

This code created a grouped bar chart, where each group represents a manufacturer, and each bar within a group represents different vehicle types.

The chart is customized with titles, axis labels, and a legend for clairy. The x axis labels manufacture names are rotated for better readability
